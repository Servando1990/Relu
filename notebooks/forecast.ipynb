```python
# Extracting the necessary columns for creating 'daily_sales'
# Assuming 'created_at' represents the date of the sale and 'line_items' contains the quantity of items sold

# The 'line_items' column seems to contain JSON-like data, so we need to parse it
import json

# Function to parse the 'line_items' column and sum the quantities
def parse_line_items(line_items_str):
    try:
        line_items = json.loads(line_items_str.replace("'", "\""))
        return sum(item['quantity'] for item in line_items)
    except:
        return 0

# Apply the function to each row in the dataframe
orders_data['total_items'] = orders_data['line_items'].apply(parse_line_items)

# Convert 'created_at' to datetime and extract the date
orders_data['created_at'] = pd.to_datetime(orders_data['created_at']).dt.date

# Group by 'created_at' and sum the 'total_items' to get 'daily_sales'
daily_sales = orders_data.groupby('created_at')['total_items'].sum().reset_index(name='daily_sales')

daily_sales.head()
```

```python
# mlforecast demo

import random

import pandas as pd
from datasetsforecast.m4 import M4
from datasetsforecast.m5 import M5
from utilsforecast.plotting import plot_series
```

```python
await M4.async_download('data', group='Hourly')
df, *_ = M4.load('data', 'Hourly')
uids = df['unique_id'].unique()
random.seed(0)
sample_uids = random.choices(uids, k=4)
df = df[df['unique_id'].isin(sample_uids)].reset_index(drop=True)
df['ds'] = df['ds'].astype('int64')

```

```python
df.head()
```

```python
fig = plot_series(df, max_insample_length=24 * 14)
fig
```

```python
from mlforecast import MLForecast
from mlforecast.target_transforms import Differences
```

```python
fcst = MLForecast(
    models=[],  # we're not interested in modeling yet
    freq=1,  # our series have integer timestamps, so we'll just add 1 in every timestep
    target_transforms=[Differences([24])],
)
prep = fcst.preprocess(df)
prep.head()
```

```python
fig = plot_series(prep)
fig
```

```python
fcst = MLForecast(
    models=[],
    freq=1,
    lags=[1, 24],
    target_transforms=[Differences([24])],    
)
prep = fcst.preprocess(df)
prep.head()
```

```python
prep.drop(columns=['unique_id', 'ds']).corr()['y']
```

```python
from numba import njit
from window_ops.expanding import expanding_mean
from window_ops.rolling import rolling_mean

@njit
def rolling_mean_48(x):
    return rolling_mean(x, window_size=48)


fcst = MLForecast(
    models=[],
    freq=1,
    target_transforms=[Differences([24])],    
    lag_transforms={
        1: [expanding_mean],
        24: [(rolling_mean, 48), rolling_mean_48],
    },
)
prep = fcst.preprocess(df)
prep.head()
```

```python
def hour_index(times):
    return times % 24

fcst = MLForecast(
    models=[],
    freq=1,
    target_transforms=[Differences([24])],
    date_features=[hour_index],
)
fcst.preprocess(df)
```

```python
from mlforecast.target_transforms import LocalStandardScaler
fcst = MLForecast(
    models=[],
    freq=1,
    lags=[1],
    target_transforms=[LocalStandardScaler()]
)
fcst.preprocess(df)
```

```python
from sklearn.base import BaseEstimator

class Naive(BaseEstimator):
    def fit(self, X, y):
        return self

    def predict(self, X):
        return X['lag1']
```

```python
fcst = MLForecast(
    models=[Naive()],
    freq=1,
    lags=[1],
    target_transforms=[LocalStandardScaler()]
)
fcst.fit(df)
preds = fcst.predict(1)
preds
```

```python
last_vals = df.groupby('unique_id').tail(1)
last_vals
```

```python
import numpy as np
np.testing.assert_allclose(preds['Naive'], last_vals['y'])
```

```python
import lightgbm as lgb
import pandas as pd
from mlforecast import MLForecast
from mlforecast.utils import generate_daily_series, generate_prices_for_series
from window_ops.expanding import expanding_mean
from window_ops.rolling import rolling_mean
```

```python
series = generate_daily_series(
    100, equal_ends=True, n_static_features=2
).rename(columns={'static_1': 'product_id'})
series.head()
```

```python
prices_catalog = generate_prices_for_series(series)
prices_catalog.head()
```

```python
series_with_prices = series.merge(prices_catalog, how='left')
series_with_prices.head()
```

```python
fcst = MLForecast(
    models=lgb.LGBMRegressor(n_jobs=1, random_state=0, verbosity=-1),
    freq='D',
    lags=[7],
    lag_transforms={
        1: [expanding_mean],
        7: [(rolling_mean, 14)]
    },
    date_features=['dayofweek', 'month'],
    num_threads=2,
)
fcst.fit(series_with_prices, static_features=['static_0', 'product_id'])
```
