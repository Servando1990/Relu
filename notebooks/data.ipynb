```python
#Convert to float the following columns: The cheapest price online store, % (Only online stores),The cheapest price all, % all +Marketplace), The cheapest price (Partstown), % OEM (Partstown)
#df['% (Only online stores)'] = df['% (Only online stores)'].str.replace(',', '.')

df['The cheapest price online store'] = df['The cheapest price online store'].str.replace(',', '.').astype(float)
df['% (Only online stores)'] = df['% (Only online stores)'].str.replace(',', '.').astype(float)
df['The cheapest price all'] = df['The cheapest price all'].str.replace(',', '.').astype(float)
df['% all +Marketplace)'] = df['% all +Marketplace)'].str.replace(',', '.').astype(float)
df['The cheapest price (Partstown)'] = df['The cheapest price (Partstown)'].str.replace(',', '.').astype(float)
df['% OEM (Partstown)'] = df['% OEM (Partstown)'].str.replace(',', '.').astype(float)





      
       
```

```python
import numpy as np

def convert_to_float(value):
    if isinstance(value, float):
        return value
    try:
        return float(value.replace(',', '.'))
    except ValueError:
        return np.nan

df['The cheapest price online store'] = df['The cheapest price online store'].apply(convert_to_float)
df['% (Only online stores)'] = df['% (Only online stores)'].apply(convert_to_float)
df['The cheapest price all'] = df['The cheapest price all'].apply(convert_to_float)
df['% all +Marketplace)'] = df['% all +Marketplace)'].apply(convert_to_float)
df['The cheapest price (Partstown)'] = df['The cheapest price (Partstown)'].apply(convert_to_float)
df['% OEM (Partstown)'] = df['% OEM (Partstown)'].apply(convert_to_float)


```
